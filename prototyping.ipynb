```python
class God():
    
    def __init__(self, name=None, aspects=None, relations=None):
        
        self.name = generate_name() if name is None else name
        
        # pick a random colour for the name of the god in markdown
        colour = '#{:02x}{:02x}{:02x}'.format(random.choice(range(256)), random.choice(range(256)), random.choice(range(256)))
        self.name_md = f"<b><font color='{colour}'>{self.name}</font></b>"
        
        # What they are the god of
        self.aspects = aspects
        
        # What relationships they have to other gods
        self.relations = relations

    @property
    def aspects_md(self):
        if len(self.aspects) == 1:
            return 'god of ' + self.aspects[0]['name'][1]
        elif len(self.aspects) > 1:
            return 'god of '+', '.join([a['name'][1] for a in self.aspects][:-1]) + ', and ' + self.aspects[-1]['name'][1]
```

```python
from IPython.display import Markdown


# Starting states of the myth
origins = [
    'single_state',
    'dual_state',
    'trinary_state',
]


pantheon_expansions = [
    {
        'name': 'dreaming',
        'number_gods': 1,
        'required_aspects': [],
    },
    {
        'name': 'vomit',
        'number_gods': 1,
        'required_aspects': [],
    },
    {
        'name': 'union',
        'number_gods': 2,
        'required_aspects': [],
    },
    # {
    #     'name': 'explosion',
    #     'number_gods': 1,
    #     'required_aspects': ['the_sun'],
    # },
]


history = []
gods = []
objects = []

# random choice of origin
origin = random.choice(origins)



# -----------IN THE BEGINNING -------------------------
if origin == 'single_state':
    first_aspect = random.choice(physical_objects)
    first_god = God(aspects=[first_aspect])
    narrative_text = f"In the beginning there was only {first_aspect['name'][1]}, {first_god.name_md}."
    history.append(narrative_text)
    gods.append(first_god)
    objects.append(first_aspect)
    
elif origin == 'dual_state':
    first_aspect, second_aspect = random.sample(physical_objects, 2)
    first_god = God(aspects=[first_aspect])
    second_god = God(aspects=[second_aspect])
    
    narrative_text = f"In the beginning there was only {first_god.name_md} {first_aspect['name'][1]}, and {second_god.name_md} {second_aspect['name'][1]}."
    history.append(narrative_text)
    gods.append(first_god)
    objects.append(first_aspect)
    gods.append(second_god)
    objects.append(second_aspect)
    
elif origin == 'trinary_state':
    first_aspect, second_aspect, third_aspect = random.sample(physical_objects, 3)
    first_god = God(aspects=[first_aspect])
    second_god = God(aspects=[second_aspect])
    third_god = God(aspects=[third_aspect])
    
    narrative_text = f"In the beginning there was {first_god.name_md} of {first_aspect['name'][1]}, {second_god.name_md} of {second_aspect['name'][1]}, and {third_god.name_md} of {third_aspect['name'][1]}."
    history.append(narrative_text)
    gods.append(first_god)
    objects.append(first_aspect)
    gods.append(second_god)
    objects.append(second_aspect)
    gods.append(third_god)
    objects.append(third_aspect)
else:
    raise NotImpelmentedError(f'Not implemented {origin}')
    
    
    
# -----------AND ON THE SECOND DAY -------------------------

# random choice of valid expansions
valid_pantheon_expansions = [
    e for e in pantheon_expansions
    if e['number_gods'] <= len(gods)
]
pantheon_expansion = random.choice(valid_pantheon_expansions)
   




    
if pantheon_expansion['name'] == 'dreaming':
    dreaming_god = random.choice(gods)
    
    num_new_gods = random.choice(range(3,7))
    new_aspects = random.sample([o for o in physical_objects + non_physical_objects if o not in objects], num_new_gods)
    new_gods = [God(aspects=[a]) for a in new_aspects]
    txt_list_of_gods = ', '.join([f'{g.name_md} {g.aspects_md}' for g in new_gods[:-1]]) + f' and {new_gods[-1].name_md} {new_gods[-1].aspects_md}'
    
    narrative_text = f"{dreaming_god.name_md} was in a deep and endless sleep. From their dreams sprang forth {txt_list_of_gods}."
    history.append(narrative_text)
    
    
if pantheon_expansion['name'] == 'vomit':
    sick_god = random.choice(gods)
    
    num_new_gods = random.choice(range(3,6))
    new_aspects = random.sample([o for o in physical_objects if o not in objects], num_new_gods)
    
    narrative_text = f"{sick_god.name_md} experienced a violent fit of nausea, and they vomited out {new_aspects[0]['name'][1]}. "
    narrative_text += f"{sick_god.name_md}'s vomiting continued, and they brought forth " + ', '.join([a['name'][1] for a in new_aspects[1:-1]])
    narrative_text += f" and finally {new_aspects[-1]['name'][1]}."
    history.append(narrative_text)
    
    
elif pantheon_expansion['name'] == 'union':
    
    partner_1, partner_2 = random.sample(gods, 2)
    
    num_new_gods = random.choice(range(3,5))
    new_aspects = random.sample([o for o in physical_objects + non_physical_objects if o not in objects], num_new_gods)
    new_aspects = random.sample([o for o in conceptual_objects if o not in objects], num_new_gods)
    new_gods = [God(aspects=[a]) for a in new_aspects]
    txt_list_of_gods = ', '.join([f'{g.name_md} {g.aspects_md}' for g in new_gods[:-1]]) + f' and {new_gods[-1].name_md} {new_gods[-1].aspects_md}'
    
    narrative_text = f"{partner_1.name_md} united with {partner_2.name_md}, and from their union were born {txt_list_of_gods}."
    history.append(narrative_text)
    
    
# else:
#     raise NotImpelmentedError(f'Not implemented {origin}')
    
    
display(Markdown('<br>'.join(history)))
```

```python
def find_siblings(target_id):
    target_objects = [o for o in all_things if o['id']==target_id]
    assert len(target_objects) == 1
    target_object = target_objects[0]
    target_parent = target_object['parent']
    
    sibling_ids = [
        o['id'] for o in all_things if o['parent']==target_parent
        if o['id']!=target_id # not self
    ]
    
    return sibling_ids

def find_children(target_id):
    children_ids = [o['id'] for o in all_things if o['parent']==target_id]
    return children_ids

find_siblings('the_land'), find_children('the_land')
```
